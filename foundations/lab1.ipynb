```python
import os
import sys
from pathlib import Path

try:
    # .py 脚本
    current_dir = Path(__file__).resolve().parent
except NameError:
    # .ipynb
    current_dir = Path.cwd()
    
project_root = current_dir.parent
sys.path.append(str(project_root))

import logging
from config.log_config import setup_logging

setup_logging()
logger = logging.getLogger(__name__)
```

```python
deepseek_api_key = os.getenv("DEEPSEEK_API_KEY")

if deepseek_api_key:
    logger.info(f"deepseek_api_key: {deepseek_api_key[:8]}")
else:
    raise RuntimeError("DEEPSEEK_API_KEY is not set")
```

```python
from openai import OpenAI

ds_ai = OpenAI(
    api_key=os.environ.get("DEEPSEEK_API_KEY"), base_url="https://api.deepseek.com"
)

messages = [{"role": "user", "content": "What is agent?"}]

resoponse = ds_ai.chat.completions.create(model="deepseek-v4-flash", messages=messages)

logger.info(f"response: {resoponse.choices[0].message.content}")
```
